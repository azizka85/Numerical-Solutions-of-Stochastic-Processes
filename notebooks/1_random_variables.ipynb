### 1. Gaussian Probability Density Function

$$
    P(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}}
$$

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-10, 10, 1000)

mu = 0
sigma = 1

y = np.exp(-0.5*(x - mu)**2/sigma**2)/np.sqrt(2*np.pi)/sigma

plt.grid(True)
plt.plot(x, y)
plt.title('$P(x) = \\frac{1}{\\sqrt{2 \\pi \\sigma^2}} e^{-\\frac{(x - \\mu)^2}{2 \\sigma^2}}$')
plt.xlabel("x")
plt.ylabel("P(x)")
plt.show()
```

### 2. Gaussian Random Variable

```python
data = np.random.randn(1000)

plt.hist(data, bins=10, density=True, alpha=0.6, color='skyblue', label='Sampled data')

x = np.linspace(-4, 4, 1000)
y = np.exp(-0.5*(x - mu)**2/sigma**2)/np.sqrt(2*np.pi)/sigma

plt.plot(x, y, 'r--', linewidth=2, label='Gaussian PDF (μ=0, σ=1)')

plt.title('Comparison of Gaussian random samples with Gaussian PDF')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.legend()
plt.grid(True)
```

### 3. Discretized Wiener Process

```python
L = 500
T = 1

dt = T/L

dW = np.sqrt(dt)*np.random.randn(L)

W = np.zeros(L)
t = np.linspace(0, T, L)

for i in range(1, L):
    W[i] = W[i - 1] + dW[i - 1]

plt.grid(True)
plt.plot(t, W)
plt.title('Discretized Wiener Process')
plt.xlabel("t")
plt.ylabel("W")
plt.show()
```
