# Implicit Methods and Numerical Stability

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 0.5
sigma = np.sqrt(6)

X0 = 1.

T = 500

ltype = ['b-', 'r--', 'm-.']

for k in range(3):
    dt = 2**-k

    N = int(T/dt)

    Xe = np.zeros(N)
    X = X0

    for i in range(N):
        dW = np.sqrt(dt)*np.random.randn()

        X = X + dt*mu*X + sigma*X*dW

        Xe[i] = np.abs(X)

    t = np.linspace(0, T, N)

    plt.semilogy(t, Xe, ltype[k], linewidth=2, label=f'∆t={dt}')

plt.grid(True)
plt.title('Discretized Brwonian Motion')
plt.xlabel("t")
plt.ylabel("X")
plt.legend()
plt.show()
```
